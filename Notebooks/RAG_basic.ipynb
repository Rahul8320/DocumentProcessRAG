```python
# Checked required env variables are set or not
import os

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY")
OPENAI_BASE_URL = os.environ.get("OPENAI_BASE_URL")

if not OPENAI_API_KEY:
    print(
        "If you are using local ollama models then ignore this warning or else set this environment value."
    )

if not OPENAI_BASE_URL:
    print(
        "If you are using local ollama models then please add this environment variable value else ignore this warning."
    )
```

```python
# Create LLM, embedding and vector store
from langchain_openai import ChatOpenAI
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma

llm = ChatOpenAI(model="qwen/qwen3-4b-2507", base_url=OPENAI_BASE_URL)
embeddings = OpenAIEmbeddings(
    model="text-embedding-embeddinggemma-300m-qat", base_url=OPENAI_BASE_URL
)
vector_store = Chroma(
    collection_name="example_collection",
    embedding_function=embeddings,
    persist_directory="./chroma_langchain_db",
)
```
